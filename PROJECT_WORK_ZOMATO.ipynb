# Step 1 - Importing libraries

```python
# Pandas is used for data manipulation and analysis
# Numpy is used for numarical analysis
# Matplotlib.pyplot and seaborn is used for data visualization
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step 2 - Create the data frame

```python
df = pd.read_csv("Zomato data .csv")
```

```python
print(df)
```

```python
df
```

# convert the data type of column - rate

```python
def handlerate(value):
  value = str(value).split('/')
  value = value[0];
  return float(value)

df['rate'] = df['rate'].apply(handlerate)
df.head()
```

```python
df.info()
```

# Type of restaurant


# que 1. What type of restraunt do the majority of customer from?

```python
df.head()
```

```python
sns.countplot(x = df['listed_in(type)']) # x liye hai x axis par
plt.xlabel('type of restaurant')
```

# conclusion - majority of the restaurant falls in dinning category

# que 2. How many votes has each type of restraunt received from customers?



```python
df.head()
```

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result,c = 'green',marker = "o")
plt.xlabel("Type of restaurant",c="red",size=20)
plt.ylabel("Votes",c="red",size=20)
```

#conclusion - dining restaurants has received maximum votes

# que 3. What are the rating that the majority of restraunt have received?

```python
df.head()
```

```python
plt.hist(df['rate'],bins=5)
plt.title("rating distribution")
plt.show()
```

#conclusion - The majorityrestaurants received ratings from 3.5 to 4

#que 4. Zomato has observed has most couples order most of their food online .what is their average spending on each order?

```python
df.head()
```

```python
couple_data = df['approx_cost(for two people)']
sns.countplot(x = couple_data)
```

# conclusion - The majority of couples prefer restaurant with an approximate cost of 300 rupees

#que 5. which mode (online or ofline) has received the maximum rating?

```python
df.head()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x = 'online_order',y = 'rate',data = df)
```

#conclusion - offline order received lower rating in comparison to online order

#que 6. which type of restaurant received more ofline orders so that zomato can provide customers with some good offers?

```python
df.head()
```

```python
pivot_table = df.pivot_table(index = 'listed_in(type)', columns = 'online_order',aggfunc='size',fill_value = 0)
sns.heatmap(pivot_table,annot = True, cmap="YlGnBu",fmt='d' )  # cmap color name hai # fmt format hai
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("listed_in(type)")
plt.show()
```

#conclusion - Dining  restaurants primarily accept offline orders,whereas cafes primarily receive online orders. This suggest that clients prefer orders in person at restaurants but prefer online ordering at cafes
